```python
# Importing essential libraries for data manipulation (pandas) and numerical operations (numpy)"
import pandas as pd
import numpy as np
```

```python
# Loading the preprocessed classification dataset from the saved CSV file
classification_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CW/classification_dataset_cleaned.csv')
```

```python
# Loading the preprocessed regression dataset from the saved CSV file
regression_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CW/regression_dataset_cleaned.csv')
```

```python
# Checking the dimensions (rows and columns) of the regression dataset
regression_data.shape
```

```python
# Displaying the first 20 rows of the regression dataset to examine its structure
regression_data.head(20)
```

```python
# Defining the specific feature columns to be used for mortality prediction
feature_cols_mortality = ['Age', 'T_Stage', 'Grade', 'Tumor_Size',
                         'Regional_Node_Examined', 'Reginol_Node_Positive', 'Survival_Months']
```

```python
# Separating features (X_mortality) and target variable (y_mortality) for mortality prediction
X_mortality = classification_data[feature_cols_mortality]
y_mortality = classification_data['Mortality_Status']
```

```python
# Importing the train_test_split function for data partitioning
from sklearn.model_selection import train_test_split
X_train_mortality, X_test_mortality, y_train_mortality, y_test_mortality = train_test_split(
    X_mortality, y_mortality, test_size=0.3, random_state=42, stratify=y_mortality)
```

```python
# Displaying information about the training and test sets and verifying class distribution
print(f"Training set shape: {X_train_mortality.shape}")
print(f"Test set shape: {X_test_mortality.shape}")
print(f"Training set class distribution:\n{y_train_mortality.value_counts()}")
print(f"Test set class distribution:\n{y_test_mortality.value_counts()}")
```

```python
# Importing different classification algorithms and voting classifier for ensemble learning
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
# Initializing the logistic regression and KNN classifiers with specific parameters
log_clf = LogisticRegression(random_state=42)
knn_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
# Creating a list of base learners (individual models) for the ensemble classifier
base_learners = [
    ('logistic', log_clf),
    ('knn', knn_clf)
]
```

```python
# Creating and training the ensemble classifier using soft voting (probability-based)
ensemble_learner = VotingClassifier(base_learners, voting='soft')
ensemble_learner.fit(X_train_mortality, y_train_mortality)
```

```python
# Making predictions on the test set using the ensemble classifier
y_pred_ensemble_learner = ensemble_learner.predict(X_test_mortality)
```

```python
# Importing various evaluation metrics for classification and regression tasks
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
```

```python
# Calculating and displaying the accuracy of the ensemble classifier
ensemble_learner_accuracy = accuracy_score(y_test_mortality, y_pred_ensemble_learner)
print("The voting ensemble classifier accuracy is: ", ensemble_learner_accuracy)
```

```python
# Importing matplotlib for visualization and plotting the confusion matrix for the ensemble model
import matplotlib.pyplot as plt 

ensemble_learner_cm_test = confusion_matrix(y_test_mortality, y_pred_ensemble_learner, labels=ensemble_learner.classes_)
ensemble_learner_disp = ConfusionMatrixDisplay(ensemble_learner_cm_test, display_labels=ensemble_learner.classes_)
ensemble_learner_disp.plot()
ensemble_learner_disp.ax_.set_title("Ensemble Learner Confusion Matrix")
plt.show()
```

```python
# Generating a detailed classification report for the ensemble classifier with precision, recall, and F1-score
print(classification_report(y_test_mortality, y_pred_ensemble_learner))
```

```python
# Plotting the ROC curve for the ensemble classifier to evaluate its performance
ensemble_roc = RocCurveDisplay.from_estimator(ensemble_learner, X_test_mortality, y_test_mortality)
plt.title("Ensemble Learner ROC Curve")
plt.grid(True)
plt.show()
```

```python
# Training the logistic regression model and evaluating its accuracy on the test set
log_clf.fit(X_train_mortality, y_train_mortality)
y_pred_log = log_clf.predict(X_test_mortality)
log_accuracy = accuracy_score(y_test_mortality, y_pred_log)
print(f"Logistic Regression accuracy: {log_accuracy:.4f}")
```

```python
# Plotting the confusion matrix for the logistic regression model
log_cm = confusion_matrix(y_test_mortality, y_pred_log, labels=log_clf.classes_)
log_disp = ConfusionMatrixDisplay(log_cm, display_labels=log_clf.classes_)
log_disp.plot()
log_disp.ax_.set_title("Logistic Regression Confusion Matrix")
plt.show()
```

```python
# Generating a detailed classification report for the logistic regression model
print(classification_report(y_test_mortality, y_pred_log))
```

```python
# Plotting the ROC curve for the logistic regression model
log_roc = RocCurveDisplay.from_estimator(log_clf, X_test_mortality, y_test_mortality)
plt.title("Logistic Regression ROC Curve")
plt.grid(True)
plt.show()
```

```python
# Training the KNN model and evaluating its accuracy on the test set
knn_clf.fit(X_train_mortality, y_train_mortality)
y_pred_knn = knn_clf.predict(X_test_mortality)
knn_accuracy = accuracy_score(y_test_mortality, y_pred_knn)
print(f"KNN accuracy: {knn_accuracy:.4f}")
```

```python
# Plotting the confusion matrix for the KNN model
knn_cm = confusion_matrix(y_test_mortality, y_pred_knn, labels=knn_clf.classes_)
knn_disp = ConfusionMatrixDisplay(knn_cm, display_labels=knn_clf.classes_)
knn_disp.plot()
knn_disp.ax_.set_title("KNN Confusion Matrix")
plt.show()
```

```python
# Generating a detailed classification report for the KNN model
print(classification_report(y_test_mortality, y_pred_knn))
```

```python
# Plotting the ROC curve for the KNN model
knn_roc = RocCurveDisplay.from_estimator(knn_clf, X_test_mortality, y_test_mortality)
plt.title("KNN ROC Curve")
plt.grid(True)
plt.show()
```

```python
# Comparing the accuracy of all three classification models
print(f"Logistic Regression accuracy: {log_accuracy:.4f}")
print(f"KNN accuracy: {knn_accuracy:.4f}")
print(f"Ensemble Learner accuracy: {ensemble_learner_accuracy:.4f}")
```

```python
# Creating a bar chart to visually compare model accuracies of the classification models
plt.figure(figsize=(10, 6))
models = ['Logistic Regression', 'KNN', 'Ensemble Learner']
accuracies = [log_accuracy, knn_accuracy, ensemble_learner_accuracy]
plt.bar(models, accuracies, color=['blue', 'green', 'red'])
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.01, f'{acc:.4f}', ha='center')
plt.show()
```

```python
# Identifying all numerical columns in the regression dataset for the survival prediction task
numerical_cols = regression_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
# Removing the target variable (Survival_Months) and Mortality_Status from the feature list
if 'Survival_Months' in numerical_cols:
    numerical_cols.remove('Survival_Months')

if 'Mortality_Status' in numerical_cols:
    numerical_cols.remove('Mortality_Status')

print("Using numerical features for survival prediction:", numerical_cols)
```

```python
# Separating features (X_survival) and target variable (y_survival) for the regression task
X_survival = regression_data[numerical_cols]
y_survival = regression_data['Survival_Months']
```

```python
# Splitting the regression data into training (70%) and testing (30%) sets
X_train_survival, X_test_survival, y_train_survival, y_test_survival = train_test_split(
    X_survival, y_survival, test_size=0.3, random_state=42)
```

```python
# Displaying information about the training and test sets for the regression task
print(f"Training set shape: {X_train_survival.shape}")
print(f"Test set shape: {X_test_survival.shape}")
```

```python
# Importing and training a Decision Tree Regressor model for survival prediction
from sklearn.tree import DecisionTreeRegressor
DT_regressor = DecisionTreeRegressor(random_state=42)
DT_regressor.fit(X_train_survival, y_train_survival)
```

```python
# Making predictions on the test set using the decision tree regressor
y_pred_survival = DT_regressor.predict(X_test_survival)
```

```python
# Evaluating the decision tree regression model using MAE, MSE, and R^2 metrics
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test_survival, y_pred_survival))
print('MSE:', metrics.mean_squared_error(y_test_survival, y_pred_survival))
print('R2:', metrics.r2_score(y_test_survival, y_pred_survival))
```

```python
# Importing tree module for decision tree visualization
from sklearn import tree
```

```python
# Visualizing the full decision tree structure (limited to top 3 levels for readability)
Tree_figure = plt.figure(figsize=(20, 10))
DT_Graph = tree.plot_tree(DT_regressor, feature_names=list(X_train_survival.columns),
                          filled=True, max_depth=3)
plt.title("Fully Grown Decision Tree (DT-1) - Top 3 Levels")
plt.tight_layout()
plt.show()
```

```python
# Creating a pruned decision tree regressor with a maximum depth of 4 to prevent overfitting
pruned_DT_regressor = DecisionTreeRegressor(max_depth=4, random_state=42)
pruned_DT_regressor.fit(X_train_survival, y_train_survival)
y_pred_pruned = pruned_DT_regressor.predict(X_test_survival)
```

```python
# Evaluating the pruned decision tree model using MAE, MSE, and R^2 metrics
print('MAE:', metrics.mean_absolute_error(y_test_survival, y_pred_pruned))
print('MSE:', metrics.mean_squared_error(y_test_survival, y_pred_pruned))
print('R2:', metrics.r2_score(y_test_survival, y_pred_pruned))
```

```python
# Visualizing the pruned decision tree structure (showing all 4 levels)
Pruned_Tree_figure = plt.figure(figsize=(15, 8))
Pruned_DT_Graph = tree.plot_tree(pruned_DT_regressor, feature_names=list(X_train_survival.columns),
                                filled=True)
plt.title("Pruned Decision Tree (DT-2) - Limited to 4 Levels")
plt.tight_layout()
plt.show()
```

```python
# Storing performance metrics for both the full and pruned decision tree models
mae_full = metrics.mean_absolute_error(y_test_survival, y_pred_survival)
mse_full = metrics.mean_squared_error(y_test_survival, y_pred_survival)
r2_full = metrics.r2_score(y_test_survival, y_pred_survival)

mae_pruned = metrics.mean_absolute_error(y_test_survival, y_pred_pruned)
mse_pruned = metrics.mean_squared_error(y_test_survival, y_pred_pruned)
r2_pruned = metrics.r2_score(y_test_survival, y_pred_pruned)
```

```python
# Comparing the performance metrics of both decision tree models
print(f"Fully Grown DT - MAE: {mae_full:.4f}, MSE: {mse_full:.4f}, R2: {r2_full:.4f}")
print(f"Pruned DT - MAE: {mae_pruned:.4f}, MSE: {mse_pruned:.4f}, R2: {r2_pruned:.4f}")
```

```python
# Creating a visual comparison of all three performance metrics between the full and pruned models
plt.figure(figsize=(15, 5))

# MAE comparison
plt.subplot(1, 3, 1)
models = ['Fully Grown DT', 'Pruned DT']
mae_values = [mae_full, mae_pruned]
plt.bar(models, mae_values, color=['blue', 'green'])
plt.title('Mean Absolute Error Comparison')
plt.ylabel('MAE (lower is better)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, mae in enumerate(mae_values):
    plt.text(i, mae + 0.01, f'{mae:.4f}', ha='center')

# MSE comparison
plt.subplot(1, 3, 2)
mse_values = [mse_full, mse_pruned]
plt.bar(models, mse_values, color=['blue', 'green'])
plt.title('Mean Squared Error Comparison')
plt.ylabel('MSE (lower is better)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, mse in enumerate(mse_values):
    plt.text(i, mse + 0.01, f'{mse:.4f}', ha='center')

# R2 comparison
plt.subplot(1, 3, 3)
r2_values = [r2_full, r2_pruned]
plt.bar(models, r2_values, color=['blue', 'green'])
plt.title('R Score Comparison')
plt.ylabel('R (higher is better)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, r2 in enumerate(r2_values):
    plt.text(i, r2 + 0.01, f'{r2:.4f}', ha='center')

plt.tight_layout()
plt.show()
```
